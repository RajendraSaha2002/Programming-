```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 400)
y = x**3 - 2*x**2 + x - 1
dy = 3*x**2 - 4*x + 1  # Derivative

plt.plot(x, y, label='f(x)')
plt.plot(x, dy, label="f'(x)")
plt.title('Function and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
